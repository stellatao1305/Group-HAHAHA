```python
import jieba  
from wordcloud import WordCloud  
from PIL import Image  
import numpy as np  
import matplotlib.pyplot as plt 
```

```python
with open('/Users/tsubomikaji/Desktop/弹幕/清理data(2版/clean3.txt','r',encoding="UTF-8") as f:
    file = f.read()
```

```python
data_cut =jieba.cut(file,cut_all = False)
```

```python
stop_words = [] 
with open("/Users/tsubomikaji/Downloads/stopwords-master/cn_stopwords.txt", 'r', encoding='utf-8') as f:
    for line in f:
        if len(line)>0:
            stop_words.append(line.strip()) 
data_result = [i for i in data_cut if i not in stop_words] 
print(data_result)
```

```python
text = " ".join(data_result).replace("\n","") 
print(text)
```

```python
wc = WordCloud(

    font_path = "/System/Library/Fonts/PingFang.ttc",
    width=1000,
    height=800,
    background_color = "white",
    max_words = 300,
)
```

```python
wc.generate(text)


wc.to_file("/Users/tsubomikaji/Downloads/heitu3.jpg") 
 

plt.imshow(wc) 
plt.axis("off") 
plt.show() 
```
